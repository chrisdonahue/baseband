```python
x = [0.0, 0.5, 1.0, 0.5, 0.0, -0.5, -1.0, -0.5]
x = np.array(x * 16)
plt.plot(x)

from scipy.io.wavfile import write
x_out = (x * 32767.0).astype(np.int16)
write('tri_24.wav', 24, x_out)
```

```python
fs = 48000.0
t = np.arange(0.0, 1.0, 1.0 / fs)[:, np.newaxis]

for f, name in zip([[500], [1500, 2000, 4000, 7000, 7500]], ['sin_500_48k.wav', 'sin_rich_48k.wav']):
    f = np.array(f)[:, np.newaxis]
    xarr = np.sin(2.0 * np.pi * t * f.T)
    x = np.sum(xarr, axis=1)
    x /= f.shape[0]
    plt.figure()
    plt.plot(x)
    plt.xlim([0, 1024])
    x_out = (x * 32767.0).astype(np.int16)
    write(name, 44100, x_out)
```
